## Part 3: Analyzing Our Data

Now that we have our data organized, let's analyze it to find patterns!

### Analysis 1: Total Count

How many AI applications did we identify?

```python
# Step 8: Count total applications

total_apps = len(app_names)
print(f"Total AI applications identified: {total_apps}")
```

### Analysis 2: Count by Category

Which AI category appears most often in our daily lives?

```python
# Step 9: Count applications in each category

recognition_count = categories.count("Recognition")
prediction_count = categories.count("Prediction")
generation_count = categories.count("Generation")
recommendation_count = categories.count("Recommendation")
optimization_count = categories.count("Optimization")

print("AI Applications by Category:")
print(f"  Recognition: {recognition_count}")
print(f"  Prediction: {prediction_count}")
print(f"  Generation: {generation_count}")
print(f"  Recommendation: {recommendation_count}")
print(f"  Optimization: {optimization_count}")
```

### Analysis 3: Most Common Category

Let's identify which category we use most.

```python
# Step 10: Find the most common category

# Create a list of all the counts
counts = [recognition_count, prediction_count, generation_count,
          recommendation_count, optimization_count]

# Create a list of category names
category_names = ["Recognition", "Prediction", "Generation",
                  "Recommendation", "Optimization"]

# Find the maximum count
max_count = max(counts)

# Find which category has that count
max_index = counts.index(max_count)
most_common_category = category_names[max_index]

print(f"\nMost common category: {most_common_category}")
print(f"Number of apps: {max_count}")
```

### Analysis 4: Average Usefulness

On average, how useful are these AI applications?

```python
# Step 11: Calculate average usefulness rating

total_usefulness = sum(usefulness)
average_usefulness = total_usefulness / len(usefulness)

print(f"Average usefulness rating: {average_usefulness:.1f}/5")

# Interpret the rating
if average_usefulness >= 4.5:
    print("These AI apps are HIGHLY useful in your daily life!")
elif average_usefulness >= 3.5:
    print("These AI apps are quite useful in your daily life.")
elif average_usefulness >= 2.5:
    print("These AI apps are moderately useful.")
else:
    print("These AI apps have limited usefulness.")
```

### Analysis 5: Most and Least Useful Apps

Which specific apps are most and least useful?

```python
# Step 12: Find most and least useful applications

# Find the highest usefulness rating
max_usefulness = max(usefulness)
max_usefulness_index = usefulness.index(max_usefulness)
most_useful_app = app_names[max_usefulness_index]

# Find the lowest usefulness rating
min_usefulness = min(usefulness)
min_usefulness_index = usefulness.index(min_usefulness)
least_useful_app = app_names[min_usefulness_index]

print(f"Most useful app: {most_useful_app} ({max_usefulness}/5)")
print(f"Least useful app: {least_useful_app} ({min_usefulness}/5)")
```

### Analysis 6: Privacy Concerns

How many of these apps raise privacy or ethical concerns?

```python
# Step 13: Count apps with privacy concerns

apps_with_concerns = has_concerns.count(True)
apps_without_concerns = has_concerns.count(False)

percent_with_concerns = (apps_with_concerns / total_apps) * 100

print(f"Apps with concerns: {apps_with_concerns} ({percent_with_concerns:.0f}%)")
print(f"Apps without concerns: {apps_without_concerns}")

# List which apps have concerns
print("\nApps that raise concerns:")
for i in range(len(app_names)):
    if has_concerns[i]:
        print(f"  - {app_names[i]}")
```

### Analysis 7: Category with Most Concerns

Which AI category tends to raise the most privacy concerns?

```python
# Step 14: Find which category has most concerns

# Count concerns by category
category_concerns = {}

for i in range(len(app_names)):
    category = categories[i]

    # If we haven't seen this category yet, add it
    if category not in category_concerns:
        category_concerns[category] = 0

    # If this app has concerns, count it
    if has_concerns[i]:
        category_concerns[category] += 1

print("Privacy concerns by category:")
for category, count in category_concerns.items():
    print(f"  {category}: {count} apps with concerns")
```
