### Overview

In this notebook, we'll analyze a week of temperature data from a smart thermostat. This demonstrates how IoT (Internet of Things) devices collect data and how we can extract meaningful insights from that data.

Imagine you have a week of temperature data from a smart thermostat. Let's analyze it to find patterns about home usage, energy consumption, and daily routines.

### Step 1: Import Necessary Libraries

We'll use two Python libraries:
- **pandas**: For organizing and analyzing data
- **datetime**: For working with dates and times

````python
# Import necessary libraries
import pandas as pd
from datetime import datetime, timedelta

print("Libraries imported successfully!")
````

### Step 2: Create Sample Data

In a real scenario, this data would come from the thermostat's API or a CSV file. For learning purposes, we'll simulate realistic smart home data.

We're simulating:
- **168 hours** (7 days Ã— 24 hours)
- **Hourly readings** of temperature
- **Occupancy status** (home = 1, away = 0)
- **Realistic patterns** (cooler at night, warmer during day, away during weekday work hours)

````python
# Generate 7 days of hourly temperature readings
dates = []
temperatures = []
occupancy = []

# Starting date: January 8, 2024 at midnight
start_date = datetime(2024, 1, 8, 0, 0)

# Simulate one week (168 hours)
for hour in range(168):
    current_time = start_date + timedelta(hours=hour)
    dates.append(current_time)

    # Simulate temperature pattern based on time of day
    # Cooler at night (68-70Â°F), warmer during day (72-76Â°F)
    hour_of_day = current_time.hour

    if 0 <= hour_of_day < 6:  # Night (12 AM - 6 AM)
        temp = 68 + (hour_of_day * 0.3)  # Gradually warming
        occupied = 1  # Home sleeping
    elif 6 <= hour_of_day < 9:  # Morning (6 AM - 9 AM)
        temp = 70 + (hour_of_day - 6) * 1.5  # Warming up
        occupied = 1  # Home, getting ready
    elif 9 <= hour_of_day < 17:  # Daytime (9 AM - 5 PM)
        temp = 74 + ((hour_of_day - 13) ** 2) * 0.1  # Varies with peak around 1 PM
        occupied = 0 if current_time.weekday() < 5 else 1  # Away on weekdays
    else:  # Evening (5 PM - midnight)
        temp = 73 - ((hour_of_day - 17) * 0.5)  # Cooling down
        occupied = 1  # Home

    temperatures.append(round(temp, 1))
    occupancy.append(occupied)

print(f"Generated {len(dates)} hours of data")
print(f"Date range: {dates[0]} to {dates[-1]}")
````

### Step 3: Create a DataFrame

A **DataFrame** is like a spreadsheet in Python. It organizes our data into rows and columns, making it easy to analyze.

Our DataFrame has three columns:
- **timestamp**: When the reading was taken
- **temperature**: Temperature in Â°F
- **occupied**: Whether someone was home (1) or away (0)

````python
# Create a DataFrame (like a spreadsheet in Python)
df = pd.DataFrame({
    'timestamp': dates,
    'temperature': temperatures,
    'occupied': occupancy
})

# Display the first few rows
print("First 10 rows of our data:\n")
print(df.head(10))
````

### Step 4: Basic Analysis - Dataset Overview

Let's start by understanding what data we have.

````python
print("=== Smart Home Temperature Analysis ===")
print("\nDataset Overview:")
print(f"Total readings: {len(df)}")
print(f"Date range: {df['timestamp'].min()} to {df['timestamp'].max()}")
````

### Step 5: Temperature Statistics

Now let's calculate basic statistics about the temperature readings.

````python
print("\nTemperature Statistics:")
print(f"Average temperature: {df['temperature'].mean():.1f}Â°F")
print(f"Minimum temperature: {df['temperature'].min():.1f}Â°F")
print(f"Maximum temperature: {df['temperature'].max():.1f}Â°F")
print(f"Temperature range: {df['temperature'].max() - df['temperature'].min():.1f}Â°F")
````

### Step 6: Occupancy Analysis

Let's compare temperature patterns when the home is occupied vs. unoccupied. This reveals interesting insights about energy usage.

````python
# Compare occupied vs. unoccupied temperatures
occupied_avg = df[df['occupied'] == 1]['temperature'].mean()
unoccupied_avg = df[df['occupied'] == 0]['temperature'].mean()

print("\nOccupancy Analysis:")
print(f"Average temp when occupied: {occupied_avg:.1f}Â°F")
print(f"Average temp when unoccupied: {unoccupied_avg:.1f}Â°F")
print(f"Difference: {abs(occupied_avg - unoccupied_avg):.1f}Â°F")
````

### Step 7: Usage Patterns

How much time is the home occupied vs. empty?

````python
print("\nUsage Patterns:")
hours_occupied = df['occupied'].sum()
hours_unoccupied = len(df) - hours_occupied

print(f"Hours home: {hours_occupied} ({hours_occupied/len(df)*100:.1f}%)")
print(f"Hours away: {hours_unoccupied} ({hours_unoccupied/len(df)*100:.1f}%)")
````

### Complete Summary Output

Let's run all the analysis together to see the complete picture:

````python
print("="*50)
print("COMPLETE SMART HOME TEMPERATURE ANALYSIS")
print("="*50)

print("\nðŸ“Š DATASET OVERVIEW")
print(f"   Total readings: {len(df)}")
print(f"   Date range: {df['timestamp'].min().strftime('%Y-%m-%d')} to {df['timestamp'].max().strftime('%Y-%m-%d')}")

print("\nðŸŒ¡ï¸  TEMPERATURE STATISTICS")
print(f"   Average temperature: {df['temperature'].mean():.1f}Â°F")
print(f"   Minimum temperature: {df['temperature'].min():.1f}Â°F")
print(f"   Maximum temperature: {df['temperature'].max():.1f}Â°F")
print(f"   Temperature range: {df['temperature'].max() - df['temperature'].min():.1f}Â°F")

print("\nðŸ  OCCUPANCY ANALYSIS")
print(f"   Average temp when occupied: {occupied_avg:.1f}Â°F")
print(f"   Average temp when unoccupied: {unoccupied_avg:.1f}Â°F")
print(f"   Difference: {abs(occupied_avg - unoccupied_avg):.1f}Â°F")

print("\nâ° USAGE PATTERNS")
print(f"   Hours home: {hours_occupied} ({hours_occupied/len(df)*100:.1f}%)")
print(f"   Hours away: {hours_unoccupied} ({hours_unoccupied/len(df)*100:.1f}%)")

print("\n" + "="*50)
````

**Why?** During weekdays (9 AM - 5 PM), the simulation assumes the AC isn't adjusted for an empty home. The temperature rises naturally. This represents a common energy inefficiencyâ€”cooling an empty home.
